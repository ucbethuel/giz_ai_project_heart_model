## Converting Data Values

```python
# Checking the distinct value of each non-numerical columns
print(f"""
Sex: {df.Sex.unique()}

ChestPainType: {df.ChestPainType.unique()}

RestingECG: {df.RestingECG.unique()}

ExerciseAngina: {df.ExerciseAngina.unique()}

ST_Slope: {df.ST_Slope.unique()}
""")
```

------------------------
As we can see above the Sex, ChestPainType, RestingECG, ExerciseAngina and ST_Slope have distinct value of 2, 4, 3, 2, and 3 respectively. so now we will map each value to a numerical value.


## Mapping non-numerical to numerical

```python
# Mapping the Sex
df.Sex = df.Sex.replace({"M": 0,"F": 1})
```

```python
# Mapping the ChestPainType
df.ChestPainType = df.ChestPainType.replace({"ATA":0,"NAP":1,"ASY":2,"TA":3})
```

```python
# Mapping RestingECG
df.RestingECG = df.RestingECG.replace({"Normal": 0,"ST": 1,"LVH": 2})
```

```python
# Mapping ExerciseAngina
df.ExerciseAngina = df.ExerciseAngina.replace({"N": 0,"Y": 1})
```

```python
# Mapping ST_Slope
df.ST_Slope = df.ST_Slope.replace({"Up": 0,"Flat": 1,"Down": 2})
```

```python
# Checking the distinct value of each non-numerical columns
print(f"""
Sex: {df.Sex.unique()}

ChestPainType: {df.ChestPainType.unique()}

RestingECG: {df.RestingECG.unique()}

ExerciseAngina: {df.ExerciseAngina.unique()}

ST_Slope: {df.ST_Slope.unique()}
""")
```

-------------------------
Alright now we have let us check the record of our dataframe again before we start building our model

```python
df.head()
```

--------------------------
Now that we have affirm that our dataset are now all in numerical values, let us separate it into features (X) and lables (y).

Where all features (X) = `df.drop("HeartDisease")`
and label (y) = `df.HeartDisease`

```python
# Assiging our feature and label to X and y respectively.
X = df.drop(columns = "HeartDisease")
y = df.HeartDisease
```

```python
# Checking the shape of X and y
X.shape, y.shape
```

-----------------------------------
## Splitting Dataset
Now let us split our data into training and testing set of 4:1 where training set has 80% and testig with 20%

But before doing that let us import our `tarin_test_split` from `sklearn.model_selection`


```python
# Importing train_test_split module
from sklearn.model_selection import train_test_split

# spliting dataset set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)
```

```python
[X_train.shape, y_train.shape], [X_test.shape, y_test.shape]
```

------------------------------------
## Building our Model
Since we are buiding a model that works on classification we will be working around with three model class which are:
* LinearRegression
* RandomForestClassifier
* DecisionTreeRegressor

After that we check accuracy score of each of them, then select the one our model best fit in.

```python
# importing the model class and sklearn.metrics
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
# Instantiating and training
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
# Testing and checking accuracy_score
y_preds = linear_model.predict(X_test)
print(f"Accuracy: {metrics.r2_score(y_test, y_preds)*100: .2f}%")
```

```python
# Instantiating and training
tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)
```

```python
# Testing and checking accuracy_score
tree_y_preds = tree_model.predict(X_test)
print(f"Accuracy: {metrics.accuracy_score(y_test, tree_y_preds)*100: .2f}%")
```

```python
# Instantiating and training
forest_model = RandomForestClassifier()
forest_model.fit(X_train, y_train)
```

```python
# Testing and checking accuracy_score
forest_y_preds = forest_model.predict(X_test)
print(f"Accuracy: {metrics.accuracy_score(y_test, forest_y_preds)*100: .2f}%")
```

## Conclusion

After building our model, it was discovered that our model performed well when build with `RandomForestClassifier`
